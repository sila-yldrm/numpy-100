### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo["item_price"]=chipo['item_price'].str.split('$').str[1]
chipo["item_price"]=chipo["item_price"].astype(float)
chipo["item_price"]
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo["item_price"].fillna(0,inplace=True)   #NaN değerler olduu için integera çeviremiyorduk
chipo['item_price'] = chipo['item_price'].apply(lambda x: int(x))
```

#### Step 13.c. Check the item price type

```python
chipo["item_price"].dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['item_price'] * chipo['quantity']).sum()
print(revenue)
```

### Step 15. How many orders were made in the period?

```python
chipo["order_id"].nunique()
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1
round(chipo.groupby("order_id")["item_price"].sum().mean(), 2)
```

```python
# Solution 2
total_revenue = chipo['item_price'].sum()
unique_orders = chipo['order_id'].nunique()
total_revenue / unique_orders
```
